```python
# load the libraries needed

import pandas as pd #for data analysis
import requests #for web urls
from bs4 import BeautifulSoup #for webscrape
```

```python
page = requests.get("http://www.supercarworld.com/cgi-bin/listall.cgi")
```

```python
soup = BeautifulSoup(page.content,"html.parser")
```

```python
carids = soup.find_all("input",{"type":"checkbox"})
```

```python
car_ids = []

for id in carids:
    x = str(id)
    car_ids.append(x[13:16])
```

```python
base_url = "http://www.supercarworld.com/cgi-bin/showgeneral.cgi?"
```

```python
counter = 0 
car_brand = []
car_name = []
num_built = []
car_class_list = []
engine_place_list = []
bhp_list = []
torque_list = []
ptw_list = []
ttw_list = []
topspeed_list = []
to60_list = []
weight_list = []
used_price_list = []
price_list = []
performance_list = []
handling_list = []
looks_list = []
features_list = []
quality_list = []
costs_list = []
desire_list = []
eco_list = []
overall_rating_list = []
money_no_obj_rating_list = []
bargain_rating_list = []

for id in car_ids:
    page = requests.get(base_url+id)
    soup = BeautifulSoup(page.content,"html.parser")
    name = soup.find("td",{"class":"heading"})
    
    number_built = soup.find_all("td", {"width":"55%"})[4]
    
    car_class = soup.find_all("td", {"width":"55%"})[0]
    
    engine_place = soup.find_all("td", {"width":"55%"})[2]
    
    bhp = soup.find_all("div", {"id":"powerI"})
    
    torque = soup.find_all("div", {"id":"torqueI"})
    
    ptw = soup.find_all("div", {"id":"ptwI"})
    
    ttw = soup.find_all("div", {"id":"ttwI"})
    
    topspeed = soup.find_all("div", {"id":"topspeedI"})
    
    to60 = soup.find_all("div", {"id":"to60I"})
    
    weight = soup.find_all("div", {"id":"weightI"})
    
    used_price = soup.find_all("td", {"width":"55%"})[-1]

    price = soup.find_all("td", {"width":"55%"})[-2]
    
    performance = soup.find_all("td", {"width":"30%"})
    if performance is not None and len(performance) > 0:
        performance = performance[0]
    
    handling = soup.find_all("td", {"width":"30%"})
    if handling is not None and len(handling) > 0:
        handling = handling[1]
    
    looks = soup.find_all("td", {"width":"30%"})
    if looks is not None and len(looks) > 0:
        looks = looks[2]
    
    features = soup.find_all("td", {"width":"30%"})
    if features is not None and len(features) > 0:
        features = features[3]
    
    quality = soup.find_all("td", {"width":"30%"})
    if quality is not None and len(quality) > 0:
        quality = quality[4]
    
    costs = soup.find_all("td", {"width":"30%"})
    if costs is not None and len(costs) > 0:
        costs = costs[5]
    
    desire = soup.find_all("td", {"width":"30%"})
    if desire is not None and len(desire) > 0:
        desire = desire[6]
    
    eco = soup.find_all("td", {"width":"30%"})
    if eco is not None and len(eco) > 0:
        eco = eco[7]
    
    overall_rating = soup.find_all("td", {"width":"25%"})
    if overall_rating is not None and len(overall_rating) > 0:
        overall_rating = overall_rating[1]
    
    money_no_obj_rating = soup.find_all("td", {"width":"25%"})
    if money_no_obj_rating is not None and len(money_no_obj_rating) > 0:
        money_no_obj_rating = money_no_obj_rating[2]
    
    bargain_rating = soup.find_all("td", {"width":"25%"})
    if bargain_rating is not None and len(bargain_rating) > 0:
        bargain_rating = bargain_rating[3]

    
    car_name.append(name)
    num_built.append(number_built)
    car_class_list.append(car_class)
    engine_place_list.append(engine_place)
    bhp_list.append(bhp)
    torque_list.append(torque)
    ptw_list.append(ptw)
    ttw_list.append(ttw)
    topspeed_list.append(topspeed)
    to60_list.append(to60)
    weight_list.append(weight)
    used_price_list.append(used_price)
    price_list.append(price)
    performance_list.append(performance)
    handling_list.append(handling)
    looks_list.append(looks)
    features_list.append(features)
    quality_list.append(quality)
    costs_list.append(costs)
    desire_list.append(desire)
    eco_list.append(eco)
    overall_rating_list.append(overall_rating)
    money_no_obj_rating_list.append(money_no_obj_rating)
    bargain_rating_list.append(bargain_rating)
    
    counter +=1
    print(counter)
    print(id)
```

```python
# need to split the dates and the car names from the car_name BS4 tag
car_name_clean = []
car_date = []
for name in car_name:
    x = name.text
    if " - " in x:
        y = x[:-11]
        z = x[-11:]
    else:
        y = x[:-4]
        z = x[-4:]
    car_name_clean.append(y)
    car_date.append(z)
```

```python
#strip out text from number built tag and add to list
num_built_clean = []
for num in num_built:
    num_built_clean.append(num.text.replace("\xa0",""))

```

```python
car_class_list_clean = []
for carclass in car_class_list:
    car_class_list_clean.append(carclass.text)
```

```python
engine_position = []
wheel_drive = []

for i in engine_place_list:
    a = i.contents
    if len(a) == 3:
        ep = a[0]
        wd = a[2]
    elif len(a) == 2:
        if any('Drive' in b for b in a):
            ep = 'blank'
            wd = a[1]
        else:
            ep = a[0]
            wd = 'blank'
    else:
        ep = 'blank'
        wd = 'blank'
    engine_position.append(ep)
    wheel_drive.append(wd)
```

```python
bhp_list_clean = []
for bhp in bhp_list:
    hp = bhp[0].text.replace("\xa0","")
    bhp_list_clean.append(hp)
    
```

```python
torque_list_clean = []
for t in torque_list:
    tt = t[0].text.replace("\xa0","")
    torque_list_clean.append(tt)
```

```python
ptw_list_clean = []
for pt in ptw_list:
    ptw = pt[0].text.replace("\xa0","")
    ptw_list_clean.append(ptw)
```

```python
ttw_list_clean = []
for tt in ttw_list:
    ttw = tt[0].text.replace("\xa0","")
    ttw_list_clean.append(ttw)
```

```python
topspeed_list_clean = []
for ts in topspeed_list:
    tsl = ts[0].text.replace("\xa0","")
    topspeed_list_clean.append(tsl)
```

```python
to60_list_clean = []
for i in to60_list:
    j = i[0].text.replace(" sec","").replace("\xa0","")
    to60_list_clean.append(j)
```

```python
weight_list_clean = []
for w in weight_list:
    x = w[0].text.replace("\xa0","").replace(" lbs","")
    weight_list_clean.append(x)
```

```python
used_price_list_clean = []
for a in used_price_list:
    b = a.text
    used_price_list_clean.append(b)
```

```python
price_list_clean = []
for p in price_list:
    q = p.text
    price_list_clean.append(q)
```

```python
performance_list_clean = []
for p in performance_list:
    if len(p) == 1:
        q= p.text.replace("/10","")
    else:
        q = ""
    performance_list_clean.append(q)
```

```python
handling_list_clean = []
for h in handling_list:
    if len(h) == 1:
        q= h.text.replace("/10","")
    else:
        q = ""
    handling_list_clean.append(q)
```

```python
looks_list_clean = []
for l in looks_list:
    if len(l) == 1:
        m = l.text.replace("/10","")
    else:
        m = ""
    looks_list_clean.append(m)
```

```python
features_list_clean = []
for f in features_list:
    if len(f) == 1:
        g = f.text.replace("/10","")
    else:
        g = ""
    features_list_clean.append(g)
```

```python
quality_list_clean = []
for q in quality_list:
    if len(q) == 1:
        r = q.text.replace("/10","")
    else:
        r = ""
    quality_list_clean.append(r)
```

```python
costs_list_clean = []
for c in costs_list:
    if len(c) == 1:
        d = c.text.replace("/10","")
    else:
        d = ""
    costs_list_clean.append(d)
```

```python
desire_list_clean = []
for d in desire_list:
    if len(d) == 1:
        e = d.text.replace("/10","")
    else:
        e = ""
    desire_list_clean.append(e)
```

```python
eco_list_clean = []
for e in eco_list:
    if len(e) == 1:
        f = e.text.replace("/10","")
    else:
        f = ""
    eco_list_clean.append(f)
```

```python
overall_rating_list_clean = []
for rating in overall_rating_list:
    if len(rating) != 0:
        orl = rating.text[20:23]
    else:
        orl = ""
    overall_rating_list_clean.append(orl)
```

```python
money_no_obj_rating_list_clean = []
for mno in money_no_obj_rating_list:
    if len(mno) != 0:
        out = mno.text[21:24]
    else:
        out = ""
    money_no_obj_rating_list_clean.append(out)
```

```python
bargain_rating_list_clean = []
for barg in bargain_rating_list:
    if len(barg) != 0:
        bargain = barg.text[20:23].replace("\n","")
    else:
        bargain = ""
    bargain_rating_list_clean.append(bargain)
```

```python
data = pd.DataFrame([car_name_clean, car_class_list_clean, num_built_clean,\
           engine_position, wheel_drive, bhp_list_clean, torque_list_clean,\
           ptw_list_clean, ttw_list_clean, topspeed_list_clean, to60_list_clean,\
           weight_list_clean, used_price_list_clean, price_list_clean,\
           performance_list_clean, handling_list_clean, looks_list_clean,\
           features_list_clean, quality_list_clean, costs_list_clean,\
           desire_list_clean, eco_list_clean, overall_rating_list_clean,\
           money_no_obj_rating_list_clean, bargain_rating_list_clean]).T
```

```python
headers = ['name', 'class', 'number_built',\
           'engine_pos', 'wheel_drive', 'bhp', 'torque',\
           'ptw', 'ttw', 'top_speed', 'to60',\
           'weight', 'used_price', 'price',\
           'performance', 'handling', 'looks',\
           'features', 'quality', 'cost',\
           'desire', 'eco', 'overall',\
           'money_no_obj', 'bargain']
```

```python
data.columns = headers
```

```python
import re
car_brand = []
for id in car_ids:
    page = requests.get(base_url+id)
    soup = BeautifulSoup(page.content,"html.parser")
    carbrand = soup.find("td",{"width":"140"})
    carbrand = str(carbrand)
    result = re.search('overview.cgi?(.*)"><img', carbrand)
    result2 = result.group(1)
    car_brand.append(result2)
```

```python
f = lambda x : x.replace("?","")
car_brand_list = list(map(f, car_brand))
```

```python
data['car_brand'] = car_brand_list
```

```python
data = data[[ 'car_brand','name', 'class', 'number_built', 'engine_pos', 'wheel_drive', 'bhp',
       'torque', 'ptw', 'ttw', 'top_speed', 'to60', 'weight', 'used_price',
       'price', 'performance', 'handling', 'looks', 'features', 'quality',
       'cost', 'desire', 'eco', 'overall', 'money_no_obj', 'bargain']]
```
